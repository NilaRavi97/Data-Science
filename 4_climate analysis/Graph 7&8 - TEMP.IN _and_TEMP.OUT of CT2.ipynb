# TEMP.IN of CT2 and TEMP.OUT of CT2

```python
# imports
import os
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
# list excel files in the folder data
path = './data'
filenames = [file for file in os.listdir(path) if file.endswith('.xlsx')]
filenames
```

Files that needs to be fetched for graph 7 and 8 : 
CAT_11_4_CT_2_DDDDDD,CT2_IN_ZZZZZZ,CT2_OUT_ZZZZZZ

```python
# read the required sheet from the excel sheet and store it into dataframe
CT2_in_sheetname = ''
CT2_out_sheetname = ''
CAT_11_4_CT2_sheetname = ''
for i in filenames:
    x = './data/'+i
    print(x)
    excel = pd.ExcelFile(x)
    excel.sheet_names
    print(excel.sheet_names)
    # ["Sheet1", "Sheet2"]
    for n in excel.sheet_names:
        if n.startswith('CT2_in_'):
            CT2_in_sheetname = n
            if(CT2_in_sheetname!=''):      
                CT2_in_df = pd.read_excel(excel, sheet_name=CT2_in_sheetname)
        if n.startswith('CT2_out_'):
            CT2_out_sheetname = n
            if(CT2_out_sheetname!=''):
                CT2_out_df = pd.read_excel(excel, sheet_name=CT2_out_sheetname)
        if n.startswith('CAT_11_4 _CT2_'):
            CAT_11_4_CT2_sheetname = n
            if(CAT_11_4_CT2_sheetname!=''):
                CAT_11_4_CT2_df = pd.read_excel(excel, sheet_name=CAT_11_4_CT2_sheetname)
```

```python
CAT_11_4_CT2_df.columns
```

```python
# find the index of the column labels
find_index_ct2 = CT2_in_df.loc[CT2_in_df['Column1'] == 'DATUM:']
ind_ct2 = find_index_ct2.index[0]
find_index_cat_11_4 = CAT_11_4_CT2_df.loc[CAT_11_4_CT2_df['Column1'] == 'Kommentar']
ind_cat_11_4 = find_index_cat_11_4.index[0]
ind_cat_11_4
```

```python
find_index_cat_11_4
```

```python
# Select rows from the CT2_in_df dataframe starting from the index specified by the ind_ct2 variable
CT2_in_df = CT2_in_df.iloc[ind_ct2:]

# Select rows from the CT2_out_df dataframe starting from the index specified by the ind_ct2 variable
CT2_out_df = CT2_out_df.iloc[ind_ct2:]

# Select rows from the CAT_11_4_CT2_df dataframe starting from the index specified by the ind_cat_11_4 variable
CAT_11_4_CT2_df = CAT_11_4_CT2_df.iloc[ind_cat_11_4:]
print(CT2_in_df)
print(CAT_11_4_CT2_df)
```

```python
#grab the first row for the header
new_headeri_ct2 = CT2_in_df.iloc[0] 
new_headero_ct2 = CT2_out_df.iloc[0] 
new_headeri_cat_11_4 = CAT_11_4_CT2_df.iloc[0]
```

```python
#pick the subsequent rows
CT2_in_df = CT2_in_df[1:]
CT2_out_df = CT2_out_df[1:]
CAT_11_4_CT2_df = CAT_11_4_CT2_df[1:]
```

```python
#change column name as selected header
CT2_in_df.columns = new_headeri_ct2
CT2_out_df.columns = new_headero_ct2
CAT_11_4_CT2_df.columns = new_headeri_cat_11_4
```

```python
# Rename the columns
CT2_in_df.rename(columns = {'DATUM:':'Date','ZEIT:':'Time','M00: %rF':'M00','M01: °C':'M01'}, inplace = True)
CT2_out_df.rename(columns = {'DATUM:':'Date','ZEIT:':'Time','M00: %rF':'M00','M01: °C':'M01'}, inplace = True)
CAT_11_4_CT2_df.rename(columns = {'Kommentar':'Date',np.nan:'Time'}, inplace = True)
```

```python
CAT_11_4_CT2_df.columns
```

```python
# change the commas with dots
CT2_in_df = CT2_in_df.apply(lambda x: x.str.replace(',','.'))
CT2_out_df = CT2_out_df.apply(lambda x: x.str.replace(',','.'))
#CAT_11_4_CT2_df = CAT_11_4_CT2_df.apply(lambda x: x.str.replace(',','.'))
```

```python
CAT_11_4_CT2_df.head()
```

```python
#remove the unwanted row
CAT_11_4_CT2_df = CAT_11_4_CT2_df.iloc[1: , :]
CAT_11_4_CT2_df.head()
```

```python
# convert all the column values to string
CAT_11_4_CT2_df = CAT_11_4_CT2_df.astype(str)
```

```python
CAT_11_4_CT2_df = CAT_11_4_CT2_df.apply(lambda x: x.str.replace(',','.'))
CAT_11_4_CT2_df.head()
```

```python
# change the datatype of required columns
# CT2 in
CT2_in_df['M01'] = CT2_in_df['M01'].astype(float)
CT2_in_df['datetime'] = pd.to_datetime(CT2_in_df['Date'] + ' ' + CT2_in_df['Time'],dayfirst=True)
```

```python
# CT2 out
CT2_out_df['M01'] = CT2_out_df['M01'].astype(float)
CT2_out_df['datetime'] = pd.to_datetime(CT2_out_df['Date'] + ' ' + CT2_out_df['Time'],dayfirst=True)
```

```python
#CAT_11_4_CT2_
CAT_11_4_CT2_df['datetime'] = pd.to_datetime(CAT_11_4_CT2_df['Date'] + ' ' + CAT_11_4_CT2_df['Time'],dayfirst=True)
```

```python
# set datetime as the index
CT2_in_df.set_index('datetime',inplace = True)
CT2_out_df.set_index('datetime',inplace = True)
CAT_11_4_CT2_df.set_index('datetime',inplace = True)
```

# Graph 7

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=CT2_in_df.index,y=CT2_in_df.M01,mode='lines',name='Temp in of CT2'))
fig.add_trace(go.Scatter(x=CAT_11_4_CT2_df.index,y=CAT_11_4_CT2_df['air IN pf plant chamber in plenum wall'],mode='lines',name='air IN pf plant chamber in plenum wall'))
fig.add_trace(go.Scatter(x=CAT_11_4_CT2_df.index,y=CAT_11_4_CT2_df['base reheat heatsink'],mode='lines',name='base reheat heatsink'))


fig.update_xaxes(
        title_text = "Date and Time",
        title_standoff = 25)

fig.update_yaxes(
        title_text = "Temperature",
        title_standoff = 25)

fig.update_layout(
 title="Temp IN of CT2"
)

fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=CT2_out_df.index,y=CT2_out_df.M01,mode='lines',name='Temp out of CT2'))
fig.add_trace(go.Scatter(x=CAT_11_4_CT2_df.index,y=CAT_11_4_CT2_df['air OUT of plant chamber in plenum wall'],mode='lines',name='air IN pf plant chamber in plenum wall'))
fig.add_trace(go.Scatter(x=CAT_11_4_CT2_df.index,y=CAT_11_4_CT2_df['base reheat heatsink'],mode='lines',name='base reheat heatsink'))


fig.update_xaxes(
        title_text = "Date and Time",
        title_standoff = 25)

fig.update_yaxes(
        title_text = "Temperature",
        title_standoff = 25)

fig.update_layout(
 title="Temp OUT of CT2"
)

fig.show()
```
